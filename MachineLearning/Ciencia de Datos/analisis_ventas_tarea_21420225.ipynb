# Análsis de ventas con NumPy

Supongamos que tienes datos de ventas de una tienda en diferentes meses y quieres analizarlos para obtener información clave, como el total de ventas, el promedio de ventas por mes y los meses con mejor desempeño.

## Paso 1. Importar NumPy y definir los datos


```python
import numpy as np

# Datos de ventas por cada mes
ventas = np.array([2300, 2500, 2600, 2100, 2900, 3100, 2800, 2700, 3000, 3300, 3500, 3700])
# Meses del año
meses = np.array(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
                  "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"])
print(ventas)
print(meses)
```

## Paso 2. Análisis de ventas con Numpy

Aplicaremos funciones clave para extraer información útil

### 1. Calcular el total y promedio de ventas al año.

```python
# Obtener el total de las ventas y el promedio, imprimirlas
total_ventas = np.sum(ventas)
promedio_ventas = np.mean(ventas)
print(f"Total de ventas en el año: ${total_ventas:,.2f}")
print(f"Promedio de ventas por mes: ${promedio_ventas:,.2f}")
```

### 2. Identificar el mes con más y menos ventas

```python
# Mes con ventas maximas
max_ventas = np.max(ventas)
mes = np.where(ventas == max_ventas)[0]
print(f"Mes con mayores ventas: {meses[mes]} ${max_ventas:,.2f}")

# Mes con ventas minimas
min_ventas = np.min(ventas)
mes = np.where(ventas == min_ventas)[0]
print(f"Mes con menores ventas: {meses[mes]} ${min_ventas:,.2f}")


```

### 3. Determinar los meses con ventas superiores al promedio

```python
# Meses con Ventas mayores al promedio
indices_superiores = np.where(ventas > promedio_ventas)
indices_superiores

print('Ventas superiores al promedio')
for i in range(indices_superiores[0].shape[0]):  # Se usa shape[0] para iterar sobre los índices
    mes = meses[indices_superiores[0][i]]
    venta = ventas[indices_superiores[0][i]]
    print(f"{mes}: ${venta:,.2f}")
```

## 4. Aplicar un incremento del 10% en las ventas

```python
proyeccion_ventas = ventas
proyeccion_ventas[0:]=proyeccion_ventas[0:]+(proyeccion_ventas[0:]*0.1)
print(f"Ventas proyectadas para el próximo año {proyeccion_ventas[0:]}")
```

## 5. Gráficar las ventas actuales y el incremento para el próximo año

```python
# importar la libreria para gráficar matplotlib
import matplotlib.pyplot as plt

# Agregar las etiquetas
plt.figure(figsize=(10,8)) # especificando el tamaño
plt.plot(meses, ventas, marker='*', color='g', label='2024')  # Graficando las ventas
plt.plot(meses, proyeccion_ventas, marker='s', color='r', label='2025')  # Graficando la proyección

plt.legend(loc='lower right')
plt.title("Tendencias de Ventas Mensuales", color='r', size=12)
plt.xlabel("Meses", color='black', size=12)
plt.ylabel("Ventas ($)", color='black', size=12)
plt.xticks(rotation=45, color='b', ha='right')  # Aquí rotamos las etiquetas 45 grados y las alineamos a la derecha
plt.yticks( color='b')  

plt.grid(True)


plt.show()
```

## 7. Detectar anomalias con un gráfico de caja (Boxplot)

```python
# Importamos la libreria seaborn
import seaborn as sns

# Crear un gráfico de caja para detectar anomalías
sns.boxplot(data=ventas, color='#FFBF00')

# Agregar título y etiquetas
plt.title("Distribucion de Ventas (Detección de Anomalías)", size=12)
plt.ylabel("Ventas", size=12)
plt.grid(True)


# Mostrar el gráfico
plt.show()
```
